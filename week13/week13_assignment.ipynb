# #10 - 

Which state got the highest percentage of 'top box' scorings, using the `Percent of patients who gave their home health agency a rating of 9 or 10 on a scale from 0 (lowest) to 10 (highest)` field.

Ignore those rows where the data is **Not Available**

````python

field1 = 'Percent of patients who gave their home health agency a rating of 9 or 10 on a scale from 0 (lowest) to 10 (highest)'
field2 = 'Number of completed Surveys'
x = hhcaps[hhcaps[field1].str.isnumeric() & hhcaps[field2].str.isnumeric()].astype({field1: float, field2: int}) 
x['top_count'] = (x[field1]*x[field2]/100)
top_box = x.groupby('State')['top_count',field2].agg('sum')
top_box['percent'] = top_box['top_count'] / top_box[field2] * 100
top_box = top_box.sort_values(by='percent', ascending=False)
answer = top_box.index[0]
print(answer)
## Solution goes here
````

````python
assert(type(answer) == str)
assert(answer == 'MS')
````
